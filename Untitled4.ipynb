```python
# ==========================================================
# EXTRAÇÃO DE ENTIDADES E RELAÇÕES (RE)
# Construção de Grafo de Conhecimento a partir da Wikipedia
# ==========================================================

# ==============================
# 1) INSTALAR DEPENDÊNCIAS
# ==============================

!pip -q install spacy wikipedia-api networkx matplotlib
!python -m spacy download en_core_web_sm -q


# ==============================
# 2) IMPORTS
# ==============================

import spacy
import wikipediaapi
import networkx as nx
import matplotlib.pyplot as plt
from spacy.matcher import Matcher


# ==============================
# 3) CARREGAR MODELO NLP
# ==============================

nlp = spacy.load("en_core_web_sm")


# ==============================
# 4) COLETAR TEXTO DA WIKIPEDIA
# ==============================

def fetch_wikipedia_text(topic):
    wiki = wikipediaapi.Wikipedia(
        language="en",
        user_agent="knowledge-graph-project"
    )

    page = wiki.page(topic)

    if not page.exists():
        print("Página não encontrada.")
        return ""

    return page.text[:5000]  # limitar tamanho para performance


# Exemplo de tema
topic = "Donald Trump"
text = fetch_wikipedia_text(topic)

print(f"Texto coletado sobre: {topic}")
print("Tamanho do texto:", len(text))


# ==============================
# 5) EXTRAÇÃO DE ENTIDADES (NER)
# ==============================

def extract_entities(text):
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return doc, entities

doc, entities = extract_entities(text)

print("\nEntidades encontradas:")
print(entities[:10])


# ==============================
# 6) EXTRAÇÃO DE RELAÇÕES (BASEADO EM PADRÃO)
# Padrão: ENTIDADE + VERBO + ENTIDADE
# ==============================

def extract_relations(doc):
    relations = []

    for sent in doc.sents:
        sent_doc = nlp(sent.text)
        ents = [ent for ent in sent_doc.ents]

        if len(ents) >= 2:
            for token in sent_doc:
                if token.pos_ == "VERB":
                    subject = None
                    obj = None

                    for child in token.children:
                        if child.dep_ in ("nsubj", "nsubjpass"):
                            subject = child.text
                        if child.dep_ in ("dobj", "attr", "pobj"):
                            obj = child.text

                    if subject and obj:
                        relations.append((subject, token.lemma_, obj))

    return relations

relations = extract_relations(doc)

print("\nRelações encontradas:")
print(relations[:10])


# ==============================
# 7) CONSTRUÇÃO DO GRAFO DE CONHECIMENTO
# ==============================

G = nx.DiGraph()

# Adicionar entidades como nós
for ent, label in entities:
    G.add_node(ent, label=label)

# Adicionar relações como arestas
for subj, verb, obj in relations:
    G.add_edge(subj, obj, relation=verb)

print("\nNós no grafo:", len(G.nodes()))
print("Relações no grafo:", len(G.edges()))


# ==============================
# 8) VISUALIZAÇÃO DO GRAFO
# ==============================

plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)

nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=1500,
    font_size=8
)

edge_labels = nx.get_edge_attributes(G, 'relation')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=6)

plt.title(f"Grafo de Conhecimento - {topic}")
plt.show()


# ==============================
# 9) LISTAR TRIPLAS EXTRAÍDAS
# ==============================

print("\nTRIPLAS (SUJEITO - RELAÇÃO - OBJETO):")
for r in relations[:20]:
    print(r)
```

```python
# ==========================================================
# RELATION EXTRACTION PIPELINE
# Dataset: ChemProt (HuggingFace)
# Modelo: SciBERT
# ==========================================================

# ==============================
# 1) INSTALAR DEPENDÊNCIAS
# ==============================

!pip -q install transformers datasets evaluate scikit-learn torch networkx matplotlib


# ==============================
# 2) IMPORTS
# ==============================

import torch
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    Trainer,
    TrainingArguments
)
import evaluate
from sklearn.metrics import classification_report


# ==============================
# 3) CARREGAR DATASET (CHEMPROT)
# ==============================

dataset = load_dataset("bigbio/chemprot")

# usar subset pequeno para teste rápido
dataset = dataset["train"].select(range(1000))

print(dataset[0])


# ==============================
# 4) PREPARAR TEXTO
# ==============================

def format_example(example):
    text = example["text"]
    return {
        "text": text,
        "label": example["labels"][0] if len(example["labels"]) > 0 else 0
    }

dataset = dataset.map(format_example)


# ==============================
# 5) TOKENIZAÇÃO
# ==============================

model_name = "allenai/scibert_scivocab_uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize_function(example):
    return tokenizer(
        example["text"],
        padding="max_length",
        truncation=True,
        max_length=256
    )

dataset = dataset.map(tokenize_function, batched=True)
dataset.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])


# ==============================
# 6) MODELO
# ==============================

num_labels = len(set(dataset["label"]))

model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=num_labels
)


# ==============================
# 7) MÉTRICAS
# ==============================

metric = evaluate.load("f1")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    f1 = metric.compute(predictions=predictions, references=labels, average="macro")
    return {"f1_macro": f1["f1"]}


# ==============================
# 8) TREINAMENTO
# ==============================

training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    per_device_train_batch_size=8,
    num_train_epochs=2,
    logging_steps=50,
    save_strategy="no"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
    eval_dataset=dataset,
    compute_metrics=compute_metrics
)

trainer.train()


# ==============================
# 9) AVALIAÇÃO
# ==============================

predictions = trainer.predict(dataset)
pred_labels = np.argmax(predictions.predictions, axis=-1)

print(classification_report(dataset["label"], pred_labels))


# ==============================
# 10) CONSTRUÇÃO DE MINI GRAFO
# ==============================

G = nx.Graph()

for i in range(50):  # pequeno subconjunto
    text = dataset[i]["text"]
    label = pred_labels[i]

    G.add_node(f"Sample_{i}")
    G.add_edge(f"Sample_{i}", f"Relation_{label}")

plt.figure(figsize=(10,6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=800)
plt.title("Mini Knowledge Graph - ChemProt Relations")
plt.show()
```

```python
# ==========================================================
# RELATION EXTRACTION + KNOWLEDGE GRAPH (CHEMPROT)
# ==========================================================

!pip -q install transformers datasets torch networkx matplotlib scikit-learn

import numpy as np
import torch
import networkx as nx
import matplotlib.pyplot as plt

from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    Trainer,
    TrainingArguments
)
from sklearn.model_selection import train_test_split


# ==============================
# 1) CARREGAR DATASET
# ==============================

dataset = load_dataset("bigbio/chemprot", "chemprot_full_source")
dataset = dataset["train"]

# Filtrar apenas exemplos com relação anotada
def has_relation(example):
    return len(example["relations"]["type"]) > 0

dataset = dataset.filter(has_relation)

print("Total com relação:", len(dataset))


# ==============================
# 2) MAPEAR LABELS
# ==============================

relation_types = set()

for example in dataset:
    for rel in example["relations"]["type"]:
        relation_types.add(rel)

relation_types = list(relation_types)
label2id = {label: i for i, label in enumerate(relation_types)}
id2label = {i: label for label, i in label2id.items()}

def format_example(example):
    rel_type = example["relations"]["type"][0]

    return {
        "text": example["text"],
        "label": label2id[rel_type],
        "entities": example["entities"]
    }

dataset = dataset.map(format_example)


# ==============================
# 3) SPLIT TRAIN / TEST
# ==============================

train_indices, test_indices = train_test_split(
    list(range(len(dataset))),
    test_size=0.2,
    random_state=42
)

train_dataset = dataset.select(train_indices)
test_dataset = dataset.select(test_indices)


# ==============================
# 4) TOKENIZAÇÃO
# ==============================

model_name = "allenai/scibert_scivocab_uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize(example):
    return tokenizer(
        example["text"],
        truncation=True,
        padding="max_length",
        max_length=256
    )

train_dataset = train_dataset.map(tokenize, batched=True)
test_dataset = test_dataset.map(tokenize, batched=True)

train_dataset.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])
test_dataset.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])


# ==============================
# 5) MODELO
# ==============================

model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=len(label2id)
)


# ==============================
# 6) TREINAMENTO
# ==============================

training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=1,
    logging_steps=50,
    save_strategy="no"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset
)

trainer.train()


# ==============================
# 7) PREDIÇÃO NO TESTE
# ==============================

predictions = trainer.predict(test_dataset)
pred_labels = np.argmax(predictions.predictions, axis=1)


# ==============================
# 8) CONSTRUÇÃO AUTOMÁTICA DO KNOWLEDGE GRAPH
# ==============================

G = nx.DiGraph()

for i, example in enumerate(test_dataset):

    entities = dataset[test_indices[i]]["entities"]

    if len(entities["text"]) >= 2:
        e1 = entities["text"][0]
        e2 = entities["text"][1]
        relation = id2label[pred_labels[i]]

        G.add_node(e1)
        G.add_node(e2)
        G.add_edge(e1, e2, relation=relation)


print("Nós no grafo:", len(G.nodes()))
print("Relações no grafo:", len(G.edges()))


# ==============================
# 9) VISUALIZAÇÃO
# ==============================

plt.figure(figsize=(12,8))
pos = nx.spring_layout(G, seed=42)

nx.draw(G, pos, with_labels=True, node_size=1000, font_size=8)

edge_labels = nx.get_edge_attributes(G, 'relation')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=6)

plt.title("Knowledge Graph - ChemProt (Predições)")
plt.show()


# ==============================
# 10) EXIBIR TRIPLAS
# ==============================

print("\nTRIPLAS EXTRAÍDAS:")
for u, v, data in G.edges(data=True):
    print((u, data["relation"], v))
```
